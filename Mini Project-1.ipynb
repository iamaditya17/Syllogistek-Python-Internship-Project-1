# CREATING AN EMPLOYEE DATABASE

```python
#1.

def add_employee(emp_id, emp_name, emp_DOJ, emp_designation, emp_salary):
    input=emp_id+','+emp_name+','+emp_DOJ+','+emp_designation+','+emp_salary+'\n'
    file=open('employee_details.txt','a')
    file.writelines(input)
    file.close()
    return 'EMPLOYEE DETAILS ARE ADDED-->>THANK YOU'
#2.

def add_hr(emp_id, hr_dept, hr_role):
    y=emp_id+','+hr_dept+','+hr_role+'\n'
    file=open('hr_details.txt','a')
    file.writelines(y)
    file.close()
    return 'HR DETAILS ARE ADDED -->>THANK YOU'
#3.

def search_employee(emp_id):
    file=open('employee_details.txt','r')
    read=file.readlines()
    for i in read:
        i = i.split(',')
        if(emp_id==i[0]):
            return i
    else:
        return 'SORRY!!! EMPLOYEE ID YOU SEARCH IS NOT FOUND'
    
    file.close()
#4.

def search_hr(emp_id):
    file=open('hr_details.txt','r')
    hr=file.readlines()
    for j in hr:
        j = j.split(',')
        if(emp_id==j[0]):
            return j
    else:
        return 'SORRY!!! HR ID YOU SEARCH IS NOT FOUND'
    
    file.close()
    
__name__='main'

while(1):
    details=int(input('1-> Add employee details\n2-> Add hr details\n3-> Search employee\n4-> Search hr\n5-> Exit\n'))
    if(details==1):
        emp_id = input('Enter the employee id ->>')
        emp_name = input('Enter the employee name ->>')
        emp_DOJ = input('Enter the employee Date Of Joining ->>')
        emp_designation = input('Enter the employee designation ->>')
        emp_salary = '0'
        print(add_employee(emp_id, emp_name, emp_DOJ, emp_designation, emp_salary))
    elif(details==2):
        emp_id = input('Enter the employee id ->>')
        hr_dept = input('Enter the department of HR ->>')
        hr_role = input('Enter the role of HR ->>')
        print(add_hr(emp_id, hr_dept, hr_role))
    elif(details==3):
        emp_id = input('Enter the employee id ->>')
        print(search_employee(emp_id))
    elif(details==4):
        emp_id = input('Enter the hr id ->')
        print(search_hr(emp_id))
    else:
        print('THANKS FOR CREATING THE EMPLOYEE DATABASE!!')
```
